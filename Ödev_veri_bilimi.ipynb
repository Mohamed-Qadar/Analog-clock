```python
import matplotlib.pyplot as plt
import seaborn as sns

# Grafik boyutunu ayarla
plt.figure(figsize=(10,6))

# Barplot oluştur
sns.barplot(x=eksik_veriler.index, y=eksik_veriler.values)

# Başlık ve etiketleri ekle
plt.title("Her Sütundaki Eksik Veri Sayısı")
plt.xlabel("Sütunlar")
plt.ylabel("Eksik Veri Sayısı")

# Grafiği göster
plt.show()
```
